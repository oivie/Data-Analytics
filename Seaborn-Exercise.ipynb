## Exercises 1


```python
# Distribution plot for TotalPay
#  I'm using the dropna() method to ensure that any NaN values in these 
# columns are excluded, as sns.distplot() would not be able to handle them.

sns.distplot(df['TotalPay'].dropna())
plt.title('Distribution of TotalPay')
plt.show()
```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

# Rename the columns to match the names in the graph
iris.columns = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width', 'species']

# Create the boxplot
sns.boxplot(data=iris, orient='h', palette='pastel')

# Remove the default title that contains the AxesSubplot information
plt.title('')

# Show the plot
plt.show()
```

```python
# Write the code to show the graph below.

The plot provided is a boxplot (also known as a whisker plot) 
for each feature of the Iris dataset. This type of plot displays 
the distribution of data based on a five-number summary: 
minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
By using the orient='h' parameter, 
the boxplot will be horizontally oriented. 
he palette='pastel' parameter provides the pastel color scheme.
```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

# Create the pair plot
# kind='scatter' specifies the type of plot for the non-diagonal subplots.
# diag_kind='hist' specifies the type of plot for the diagonal subplots.
# palette='blue' gives the blue color to the plots.
sns.pairplot(iris, kind='scatter', diag_kind='hist', palette='blue')

# Show the plot
plt.show()
```

Create a Categorical Plot for the column Sex of the Titanic dataset.

```python
# Load the Titanic dataset
titanic = pd.read_csv('Titanic.csv')

# Create a categorical plot for the 'Sex' column
sns.countplot(x='Sex', data=titanic)

# Set the title and labels
plt.title('Distribution of Passengers by Sex')
plt.ylabel('Number of Passengers')

# Show the plot
plt.show()
```
